```python
import pandas as pd
```

```python
file_path = "/Users/rizqiamaliakartika/PYTHON/Portofolio/dataset/DataCoSupplyChainDataset.csv"
df = pd.read_csv(file_path, encoding='latin1')  # atau 'windows-1252'
```

```python
df.isnull().sum()
```

```python
df['order date (DateOrders)'] = pd.to_datetime(df['order date (DateOrders)'], errors='coerce')
df['shipping date (DateOrders)'] = pd.to_datetime(df['shipping date (DateOrders)'], errors='coerce')
```

```python
df['Shipping_Delay'] = (df['shipping date (DateOrders)'] - df['order date (DateOrders)']).dt.days
df['Delay_Diff'] = df['Days for shipping (real)'] - df['Days for shipment (scheduled)']
df['Is_Late'] = df['Shipping_Delay'] > df['Days for shipment (scheduled)']
```

```python
df.drop(columns=[
    'Customer Email', 'Customer Password', 'Product Image', 'Product Description', 'Customer Zipcode', 'Order Zipcode'
], inplace=True, errors='ignore')
```

```python
df.head()
```

```python
df['Customer Lname'].fillna('Unknown', inplace=True)
df['Customer Zipcode'].fillna(-1, inplace=True) 
# `Order Zipcode` banyak null, bisa dibiarkan atau dihapus jika tidak relevan
df.drop(columns=['Order Zipcode'], inplace=True)
```

```python
df['Delivery Status'] = df['Delivery Status'].str.strip().str.lower()
df['Customer Segment'] = df['Customer Segment'].str.strip().str.lower()
df['Order Status'] = df['Order Status'].str.strip().str.lower()
df['Shipping Mode'] = df['Shipping Mode'].str.strip().str.lower()
```

```python
def fix_coordinate(val):
    # Hilangkan semua titik atau koma
    clean_val = str(val).replace('.', '').replace(',', '')
    # Sisipkan titik desimal setelah 2 digit pertama
    if len(clean_val) > 2:
        return float(clean_val[:2] + '.' + clean_val[2:])
    else:
        return float(clean_val)

df['Latitude'] = df['Latitude'].apply(fix_coordinate)
df['Longitude'] = df['Longitude'].apply(fix_coordinate)
```

```python
df.to_csv("/Users/rizqiamaliakartika/PYTHON/Portofolio/dataset/data_cleaned.csv", index=False)
print("✅ File berhasil dibersihkan dan disimpan!")
```
